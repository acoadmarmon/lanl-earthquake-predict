```python
#convert to tfrecords

import tensorflow as tf
csv = pd.read_csv("../../data/train_data_reformatted.csv").values
with tf.python_io.TFRecordWriter("../../data/train_data_reformatted.tfrecords") as writer:
    for row in csv:
        features, label = row[:-1], row[-1]
        example = tf.train.Example()
        example.features.feature["features"].float_list.value.extend(features)
        example.features.feature["label"].float_list.value.append(label)
        writer.write(example.SerializeToString())
```

```python
def read_csv(seen_rows, nrows, k):
    df = pd.read_csv('../../data/train.csv', skiprows=seen_rows, nrows=nrows)
    matrix = df.values
    j = 0
    output = np.zeros((512, 15001))
    for i in range(0, nrows - 15000, 7500):
        output[j] = np.append(matrix[i:i+15000, 0], matrix[i+14999, 1])
        j = j + 1
    pd.DataFrame(output).to_csv('../../data/processed_data/train_data_reformatted' + str(k) + '.csv', header=False, index=False)
    print(k)
    k = k + 1
    seen_rows = seen_rows + nrows
```
